```python
import keras
from keras.layers import LSTM, TimeDistributed, Dense, RepeatVector, Embedding
from keras.models import Sequential
import numpy as np
```

```python
X_train = np.array([[1,3,4,8,3,6]])
Y_train = np.array([[[1],[2],[4],[10],[2]]])
```

```python
X_train, Y_train
```

```python
model = Sequential()
model.add(Embedding(10,32, input_length=6))
model.add(LSTM(10))
model.add(RepeatVector(5))

model.add(LSTM(10, return_sequences=True))
model.add(TimeDistributed(Dense(1)))
model.summary()
```

```python
model.compile(loss='mean_squared_error',  optimizer = 'adam')
model.fit(X_train, Y_train,epochs = 10000)
```

```python
np.around(model.predict(X_train))
```
