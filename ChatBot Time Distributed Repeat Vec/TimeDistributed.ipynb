```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import TimeDistributed
```

<H1> One To One LSTM Network </H1> <br>
All the LSTM always take in the input in form of a 3D vector, Hence X must have 3 Dimension in all the cases, In this One-to-One LSTM we take one example and then we predict the corresponding value, i.e. input 0.2 and output 0.2 then take in 0.4 and output 0.4 and so on so forth. Hence this is a one-to-one LSTM network.

```python
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(len(seq), 1, 1)
y = seq.reshape(len(seq), 1)

# define LSTM configuration
n_neurons = 10
n_batch = length
n_epoch = 1000
```

```python
# create LSTM
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(1, 1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
X.shape, X
```

```python
y.shape, y
```

```python
# train LSTM
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
# evaluate
result = model.predict(X[:3], batch_size=n_batch, verbose=0)
for value in result:
	print('%.1f' % value)
```

<H1> Many to One LSTM Network </H1>
<br>
In this LSTM we take in the input as all the numbers and then output the entire matrix at once as shown below
input = [[[0. ],[0.2],[0.4],[0.6],[0.8]]] and output = [[0. , 0.2, 0.4, 0.6, 0.8]] Hence the One-To-Many Architecture

```python
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length)
```

```python
# create LSTM
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(length, 1)))
model.add(Dense(length))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
X, y
```

```python
# train LSTM
model.fit(X, y, epochs=n_epoch, batch_size=1, verbose=2)
# evaluate
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result[0,:]:
	print('%.1f' % value)
```

<h1> Many to Many LSTM Network </h1> In this case we make a hybrid sort of NN where we input the entire sequence and output the entire sequence much like the second case! But Have a look at the number of parameters in the second case! They are way too many! However the number of parameters in this case is same as that in case 1 Hence this makes the NN really fast as compared to the many to many case as shown above

```python
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length, 1)
```

```python
# create LSTM
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(length, 1), return_sequences=True))
model.add(TimeDistributed(Dense(1)))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
X, y
```

```python
# train LSTM
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
# evaluate
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result[0,:,0]:
	print('%.1f' % value)
```
